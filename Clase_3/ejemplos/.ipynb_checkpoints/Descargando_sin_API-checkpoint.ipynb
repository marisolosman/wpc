# Descargando la temperatura del sito del SMN

```python
import requests

url = 'http://www.smn.gov.ar/prongraf/tabla-pronciu.php?provincia=C%F3rdoba&ciudad=R%EDo-Cuarto&var=cordoba&imagen=ZN'
html = requests.get(url)
print(html.text)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html.text, 'html.parser')
table = soup.find('td', attrs={'colspan':'2', 'align': 'center', 'bgcolor':'#EAF4FF'})
print(table)
```

```python
minima = table.find('font', attrs={'color':'#0000FF'})
maxima = table.find('font', attrs={'color':'#FF0000'})
# despues de las 18 hora, hay una sola temperatura
unica = table.find('font', attrs={'color':'#000000'})

try:
    print("mínima: {0}, máximia: {1}".format(minima.text, maxima.text))
except:
    print("temperatura: {0}".format(unica.text))
```
